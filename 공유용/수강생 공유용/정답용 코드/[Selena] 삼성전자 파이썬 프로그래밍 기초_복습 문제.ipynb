## 1) [숫자형 - 연산자] 문제

```python
# [숫자형 - 연산자]
# 1) 3의 배수인지 여부를 판별하는 조건문 만들기
# n 입력 받아서  3의 배수라면 "3의 배수입니다." 출력
#                3의 배수가 아니라면 "3의 배수가 아닙니다." 출력
# n이 1일 때, 3의 배수가 아닙니다. 출력
# n이 3일 때, 3의 배수입니다. 출력

# 나눗셈의 나머지 연산 => %
n = int(input("숫자를 입력해주세요. :"))

if n%3 == 0 :
  print('3의 배수입니다.')
else :
  print('3의 배수가 아닙니다.')
```

## 2) [숫자형 - 연산자] 문제

```python
# [숫자형 - 연산자]
# 2) 나눗셈의 몫 연산자를 사용하여 아래의 조건문 만들기
# n 입력 받아서 5로 나누었을 때 몫이 1이 나온다면 "pass" 출력
#               5로 나누었을 때 몫이 1이 나오지 않는다면 "fail" 출력
# n이 5일 때, pass 출력
# n이 10일 때, fail 출력
# 나눗셈의 몫 연산자 => //
n = int(input("숫자를 입력해주세요. :"))

if n//5 == 1 :
  print('pass')
else :
  print('fail')
```

## [논리 연산 - 단락 평가] 이론

- 논리 연산 - 단락 평가
> - 단락 평가는 첫 번째 값만으로 결과가 확실할 때 두 번째 값은 확인(평가)하지 않는 방법을 말함.
> - 즉, and 연산자는 두 값이 모두 참이라야 참이므로 첫 번째 값이 거짓이면 두 번째 값은 확인하지 않고 바로 거짓으로 결정!
> - 파이썬에서 논리 연산자는 이 단락 평가에 따라 반환하는 값이 결정

```python
# or (논리 합) : 두 개의 값이 모두 False인 경우에만 
# 결과 값으로 False 반환, 나머지는 모두 True 반환
# 첫 번째 조건인 k == 200은 k = 100 일때 False 반환
# 두 번째 조건인 300은 True 반환
# 세 번째 조건인 400은 True 반환
# 해당 조건은 (k == 200 or 300 or 400) =>  하나라도 True라면 True 반환
k = 100

if k == 200 or 300 or 400 :
    print(k)
else:
    print('종료')
```

```python
# and (논리 곱) : 두 개의 값이모두 True인 경우에만 
# 결과 값으로 True 반환, 나머지는 모두 False 반환
# 첫 번째 조건인 k == 200은 k = 100 일때 False 반환
# 두 번째 조건인 300은 True 반환
# 세 번째 조건인 400은 True 반환
# 해당 조건은 (k == 200 or 300 or 400) =>  하나라도 False라면 False 반환
k = 100

if k == 200 and 300 and 400:
    print(k)
else:
    print('종료')
```
