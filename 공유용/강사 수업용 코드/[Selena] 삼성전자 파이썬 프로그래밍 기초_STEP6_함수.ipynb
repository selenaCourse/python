### **1. 기본 구조**
- 반복적으로 사용되는 기능을 함수로 사용
- 함수는 def 키워드 사용하여 생성 







````python
# add 함수 정의
# 매개변수 x와 y를 입력으로 받기
# x + y 코드를 계산하여 출력
def add(x, y):
  return x + y
````

````python
# add 함수를 호출하여 인수로 3과 6 입력
# z값에 함수의 결과를 출력하여 저장 
z = add(3, 6)
print(z)
````

### **2. 매개변수 (parameter)**
- 함수에 입력으로 전달된 값을 받는 변수

#### **2-1. 디폴트 매개변수**
- 매개변수의 값이 입력되지 않으면 디폴트 값으로 자동 입력
- 매개변수가 여러 개인 경우 뒤의 매개변수부터 디폴트 값을 가질 수 있음



````python
# add 함수 정의
# 매개변수 x와 y를 입력으로 받기
# y는 디폴트 매개변수로 입력이 없다면 디폴트 값으로 자동 입력
# x + y 코드를 계산하여 출력
def add(x, y=3):
  return x + y
````

````python
# add 함수를 호출하여 인수로 3과 6 입력
# z값에 함수의 결과를 출력하여 저장 
# y 매개변수에 값을 입력하면 그대로 입력 
z = add(3, 6)
print(z)
````

````python
# add 함수를 호출하여 인수로 3만 입력
# z값에 함수의 결과를 출력하여 저장 
# y 매개변수에 입력이 없다면 디폴트 값으로 자동 입력 
z = add(3)
print(z)
````

### **3. 인수 (arguments)**
- 함수를 호출할 때 전달하는 입력 값

- ```
함수(인수1, 인수2)
```



````python
# add 함수 정의
# 매개변수 x와 y를 입력으로 받기
# x + y 코드를 계산하여 출력
def add(x, y):
  return x + y
````

````python
# add 함수를 호출하여 인수로 3과 6 입력
# 인수는 함수를 호출할 때 전달하는 입력 값
add(3, 6)
````

### **4. 리턴 (return)**
- 함수를 호출한 곳으로 결과 값 반환
- ```
def 함수이름(매개변수)
      실행문
      return 되돌려 주는 결과 값
```

- ```
변수 = 함수이름(인수)
```




#### **4-1. 한 개의 리턴 값**

````python
# returnTest 함수 정의하고 함수를 호출하여 return 값 확인
def returnTest() :
  return "되돌려 주는 결과 값"

value = returnTest()
print(value)
````

#### **4-2. 두 개의 리턴 값**

````python
# returnTest 함수 정의하고 함수를 호출하여 return 값 확인
# 반환되는 두 개의 값을 차례대로 value1, value2에 저장하여 출력
def returnTest() :
  a = 3
  b = 4
  return a, b

value1, value2 = returnTest()
print(value1)
print(value2)
````

## **1. 함수 문제 (1)**

* 디폴트 매개변수



````python
# 1) multiplication 함수 정의
#    - 매개변수 a, b 중에서 b는 2로 디폴트 매개변수로 정의
#    - a * b로 계산하여 출력
def multiplication(a, b=2) :
  return a * b
````

````python
# 2) multiplication 함수를 호출하여 인수로 3과 6 입력하여 c 변수에 저장하여 출력
# b 매개변수에 값을 입력하면 a=3, b=6 그대로 입력되어 함수 결과 추출 
c = multiplication(3, 6)
print(c)
````

````python
# 3) multiplication 함수를 호출하여 인수로 3 입력하여 c 변수에 저장하여 출력
# b 매개변수에 입력이 없다면 디폴트 값으로 b=2 자동 입력되어 함수 결과 추출 
c = multiplication(3)
print(c)
````
